```python
import pandas as pd
df = pd.read_csv("Wine.csv")
```

```python
X = df.drop(["Customer_Segment"],1)
```

```python
y = df["Customer_Segment"]
```

```python
X
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.model_selection import train_test_split as tts
```

```python
log_reg = LogisticRegression(random_state=42)
```

```python
X_train,X_test,y_train,y_test = tts(X,y,test_size=0.25,random_state=42)
log_reg.fit(X_train,y_train)
y_pred = log_reg.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
df = pd.read_csv("diabetes.csv")
```

```python
X = df.drop(["diabetes"],1)
y = df["diabetes"]
```

```python
X_train,X_test,y_train,y_test = tts(X,y,test_size=0.25,random_state=42)
```

```python
log_reg = LogisticRegression()
from sklearn.neighbors import KNeighborsClassifier
knn=  KNeighborsClassifier(n_neighbors=3)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier(max_depth=4,min_samples_leaf=0.12)
```

```python
model1 = dtc
model2 = knn
model3= log_reg
```

```python
model1.fit(X_train,y_train)
model2.fit(X_train,y_train)
model3.fit(X_train,y_train)
```

```python
pred1=model1.predict(X_test)
pred2=model2.predict(X_test)
pred3=model3.predict(X_test)
```

```python
import numpy as np
final_pred = np.array([])
for i in range(0,len(X_test)):
    final_pred = np.append(final_pred, mode([pred1[i], pred2[i], pred3[i]]))
```

```python
from sklearn.ensemble import VotingClassifier
model = VotingClassifier(estimators=[('dtc', model1), ('knn', model2), ('logreg', model3)], voting='soft')
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
(accuracy_score(pred1,y_test)+accuracy_score(pred1,y_test)+accuracy_score(pred1,y_test))/3
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn import tree
model = BaggingClassifier(tree.DecisionTreeClassifier(random_state=1))
model.fit(X_train, y_train)
model.score(X_test,y_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
model = AdaBoostClassifier(random_state=1)
model.fit(X_train, y_train)
model.score(X_test,y_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model= GradientBoostingClassifier(learning_rate=0.01,random_state=1)
model.fit(X_train, y_train)
model.score(X_test,y_test)
```
